```python
from llama_cpp import Llama

# define n_ctx manually to permit larger contexts
LLM = Llama(model_path="models/llama-2-7b-chat.Q8_0.gguf", n_ctx=2048)

# create a text prompt
prompt = "Q: What is the meaning of life? A:"

# set max_tokens to 0 to remove the response size limit
output = LLM(prompt, max_tokens=128)

# display the response
print(output["choices"][0]["text"])
```
